```python
#To filter how many people in india know python..here sum method all the true as 1 and false as 0 and sum all the ones
filt = df["Country"] == "India"
df.loc[filt]["LanguageWorkedWith"].str.contains("Python").sum()
```

```python
#We cannot user str attribute to series group by object..we have to use lambda function
country_grp["LanguageWorkedWith"].str.contains("Python").sum()
```

```python
country_grp["LanguageWorkedWith"].apply(lambda x: x.str.contains("Python").sum())
```

```python
#TO find per of each country knows python
country_correspondence = df["Country"].value_counts()
country_correspondence
```

```python
country_uses_python = country_grp["LanguageWorkedWith"].apply(lambda x: x.str.contains("Python").sum())
country_uses_python
```

```python
#To find percentage, using concat method going to concat both the above result
python_df = pd.concat([country_correspondence, country_uses_python], axis="columns", sort=False)
```

```python
python_df
```

```python
#renaming the columns
python_df.rename(columns={"Country":"NumRespodents","LanguageWorkedWith":"NumKnowsPython"}, inplace=True)
```

```python
python_df
```

```python
#creating a new column to find percentage

python_df["PerKnowsPython"] = (python_df["NumKnowsPython"] / python_df["NumRespodents"]) * 100
```

```python
python_df
```

```python
#sorting values using percentage column
python_df.sort_values(by="PerKnowsPython", ascending=False, inplace=True)
```
